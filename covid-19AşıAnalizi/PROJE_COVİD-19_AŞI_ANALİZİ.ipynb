```python
data["total_vaccinations"].value_counts().plot.bar()
print("Mean:{}".format(data["total_vaccinations"].mean()))
print("Median:{}".format(data["total_vaccinations"].median()))
print("Mod:{}".format(data["total_vaccinations"].mode()))
#Toplam aşılanmaya göre ortalama,mod ve medyanı bulalım.
```

```python
#daily_vaccinations_raw özniteliğinin modu,medyanı,ortalaması
data['daily_vaccinations_raw'].mode()
data['daily_vaccinations_raw'].median()
data['daily_vaccinations_raw'].mean()
```

```python
data["people_vaccinated"].value_counts().plot.bar()
print("Mean:{}".format(data["people_vaccinated"].mean()))
print("Median:{}".format(data["people_vaccinated"].median()))
print("Mod:{}".format(data["people_vaccinated"].mode()))
#Aşılanmış insanlara göre ortalama,mod ve medyanı bulalım.
```

```python
labels = ('total_vaccinations_per_hundred','people_vaccinated_per_hundred','people_fully_vaccinated_per_hundred')
sizes = [42.742,40.732,38.227 ]
colors = ['yellow','lightblue','green']
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%')
plt.title('Covid-19 Aşı Analizi')
plt.show()
#Pasta grafiğiyle çizelim.
```

```python
f=plt.figure(figsize=(12,6))

plt.bar(data.country,data.total_vaccinations,width=0.5,
        color="#f9af2b",edgecolor="#f05131",
        label="total_vaccinations")

plt.ylabel("Oran")
plt.xlabel("country")
plt.legend()
plt.title("total_vaccinations Oranı")
plt.show()
#Ülkelere göre bar grafiğiyle çizelim.
```

```python
plt.figure(figsize=(12,6))
plt.plot(data.country,data.people_fully_vaccinated,color="red") 
plt.plot(data.country,data.people_fully_vaccinated,color="blue")
plt.xlabel("country")
plt.ylabel("Toplam Aşılanmış ve Tam Aşılanmış Oranı")
plt.title("Covid-19 Toplam Aşılanmış ve Tam Aşılanmış İnsanlar Oranı")
plt.show()
#İki farklı veriyi tek grafikte gösterelim.
```

```python
# 9'a 3'lük bir alan için grafik sahası oluşturuyoruz(kutu grafiği)
plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.boxplot(data.total_vaccinations)
plt.title("total_vaccinations")

plt.subplot(132)
plt.boxplot(data.people_vaccinated)
plt.title("people_vaccinated")

plt.subplot(133)
plt.boxplot(data.people_fully_vaccinated)
plt.title("people_fully_vaccinated")

plt.show()
```
